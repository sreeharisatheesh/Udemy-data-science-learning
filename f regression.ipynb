```python
import pandas as pd
from sklearn.metrics import r2_score,accuracy_score,mean_squared_error
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("1.03.+Dummies.csv")
df
```

```python
df['Attendance'] = df['Attendance'].map({'Yes':1,"No":0})
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
model  = LinearRegression()
model.fit(df[['SAT','GPA']],df['Attendance'])
```

```python

# Predicted values
ypredi = model.predict(df[['SAT', 'GPA']])

# R2 score
r2 = r2_score(df['Attendance'], ypredi)
print(f"R2 Score: {r2}")

# Mean Squared Error
mse = mean_squared_error(df['Attendance'], ypredi)
print(f"Mean Squared Error: {mse}")
```

```python
y = df['Attendance']
x = sm.add_constant(df[['SAT', 'GPA']])
results = sm.OLS(y,x).fit()
results.summary()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("real_estate_price_size_year.csv")
df
```

```python
x = df[['size','year']]
y = df['price']
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.predict([[1000,2010]])
```

```python
model.score(x,y)
```

```python
def adj_r(x,y):
  n = x.shape[0]
  p = x.shape[1]
  r2 = model.score(x,y)
  adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
  return adjusted_r2
```

```python
adj_r(x,y)
```
